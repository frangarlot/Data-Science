## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
moving_avg=pd.DataFrame()

moving_avg['2008 BAC Close'] = bank_stocks.loc['2008']['BAC']['Close']
moving_avg['2008 BAC 30 day average'] = bank_stocks.loc['2008']['BAC']['Close'].rolling(30).mean()

moving_avg

```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=moving_avg, dashes=False, palette='viridis_r', lw=1.8)  # _r EN PALETTE ES PARA USAR LOS COLORES PERO ALREVEZ!
```

**Create a heatmap of the correlation between the stocks Close Price.**

```python
plt.figure(figsize=(12,6))
sns.heatmap(bank_stocks.xs('Close',level=1,axis=1).corr(), annot=True,cmap='coolwarm')
```

**Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(bank_stocks.xs('Close',level=1,axis=1).corr(), annot=True,cmap='coolwarm',figsize=(8,8))
```

```python
close_corr = bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
close_corr.iplot(kind='heatmap',colorscale='rdylbu')
```
